VGG16

```python
import numpy as np
import cv2
import tensorflow as tf
from tensorflow import keras
```

```python
path_train="../input/image-classification/data/train"
path_test="../input/image-classification/data/test"
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_batches=ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet.preprocess_input).flow_from_directory(directory=path_train,target_size=(224,224),batch_size=10)
test_batches=ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet.preprocess_input).flow_from_directory(directory=path_test,target_size=(224,224),batch_size=10)
```

```python
img,label=next(train_batches)
```

```python
vgg16_model=tf.keras.applications.vgg16.VGG16()
```

```python
model=keras.Sequential()
for layer in vgg16_model.layers[:-1]:
    model.add(layer)
```

```python
for layer in model.layers:
    layer.trainable=False
```

```python
model.add(keras.layers.Dense(6,activation='softmax'))
```

```python
model.compile(optimizer="Adam", loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
model.fit(train_batches,epochs=10,verbose=2)
```

```python
model.evaluate(test_batches)
```

ResNet50

```python
import matplotlib.pyplot as plt
import numpy as np
import PIL
import tensorflow as tf
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Layer,Flatten
from tensorflow.keras.optimizers import Adam
```

```python
resnet_model = Sequential()

pretrained_model= tf.keras.applications.ResNet50(include_top=False,
                   input_shape=(224,224,3),
                   pooling='avg',classes=6,
                   weights='imagenet')
for layer in pretrained_model.layers:
        layer.trainable=False

resnet_model.add(pretrained_model)
```

```python
resnet_model.add(Dense(6, activation='softmax'))
```

```python
#resnet_model.summary()
```

```python
resnet_model.compile(optimizer=Adam(lr=0.001),loss='categorical_crossentropy',metrics=['accuracy'])

history = resnet_model.fit(train_batches, epochs=10)
```
